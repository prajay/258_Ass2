```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
with open('train.tsv') as f:
    data = pd.read_csv(f, sep='\t')
```

```python
data
```

```python
len(data)
```

```python
len(data['category_name'].unique())
```

```python
len(data['brand_name'].unique())
```

```python
data['brand_name'] = data['brand_name'].fillna("NA")
data['category_name'] = data['category_name'].fillna("NA")
```

```python
data['basic_category'] = data['category_name'].apply(lambda x : x.split('/')[0])
```

```python
category_to_index = {}
for i, val in np.ndenumerate(data['basic_category'].unique()):
    category_to_index[val] = i[0]
```

```python
data['basic_category'] = data['basic_category'].apply(lambda x : category_to_index[x])
```

```python
enc = OneHotEncoder()
d = data['basic_category'].values.reshape(-1,1)
enc.fit(d)
one_hot_category = enc.transform(d).toarray()
```

```python
top_brands = data['brand_name'].value_counts()[:51]
top_brands = set(top_brands.index)
data['top_brand'] = data['brand_name'].apply(lambda x : x if x in top_brands else "NA")
```

```python
len(data['item_condition_id'].unique())
```

```python
enc = OneHotEncoder()
enc.fit(data['item_condition_id'].values.reshape(-1,1))
enc.n_values_
```

```python
enc.feature_indices_
```

```python
one_hot_condition_id = enc.transform(data['item_condition_id'].values.reshape(-1,1)).toarray()
```

```python
brand_to_index = {}
for i, val in np.ndenumerate(data['top_brand'].unique()):
    brand_to_index[val] = i[0]
```

```python
data['top_brand'] = data['top_brand'].apply(lambda x : brand_to_index[x])
```

```python
enc = OneHotEncoder()
d = data['top_brand'].values.reshape(-1,1)
enc.fit(d)
one_hot_top_brand = enc.transform(d).toarray()
```

```python
prices = data['price'].values.reshape(-1,1)
```

```python
X = np.hstack([one_hot_condition_id, one_hot_category, one_hot_top_brand, prices])
```
